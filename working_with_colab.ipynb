## Mounting Google Drive

We have some data in Google Drive that we will use for this tutorial. To access the data in colab, we need to mount Google Drive in our Colab environment.

In **Google Drive**,

- Find the data folder `Spool_example_data`.
- Right-click on the folder, and go to `organize`.
- Select `Add shortcut to Drive`. This will create a shortcut to the folder in your Google Drive.
- Now, we can mount Google Drive in our Colab environment.

In **Colab**,

- Find the `Files` tab on the left side of the screen.
- Click on the `Mount Drive` button. This will create a cell in your notebook that looks like this:

```python
from google.colab import drive
drive.mount('/content/drive')
```

- Run the cell. This will prompt you to authorize Colab to access your Google Drive.
- Once you authorize it, there should be a message "Mounted at /content/drive". This means that your Google Drive will now be accessible by the folder `/content/drive/MyDrive/`. You can prefix other file paths with this to access files in your Google Drive.
