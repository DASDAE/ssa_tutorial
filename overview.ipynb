# Overview and Expected Outcomes
## What is dascore?

DASCore is a Python package for working with DAS data. It provides a set of tools for

- reading and writing of DAS data files
- Simple processing and visualizing DAS data.

Dascore has in-built examples that we can use to test the package.

## Outline

- Installing dascore
- Reading and writing DAS data files
- Simple processing of DAS data
- Visualizing DAS data

## Reading DAS data files

Dascore provides a set of functions for reading and writing DAS data files. The main function for reading is `dascore.spool()`.

```python
import dascore as dc
spool = dc.spool("path/to/das/data/file")
```

- Example:

```python
from dascore.utils.downloader import fetch 

# Fetch a sample file path from DASCore
file_path = fetch('terra15_das_1_trimmed.hdf5')
spool = dc.spool(file_path)
```

````python
import dascore as dc
from dascore.utils.downloader import fetch 

# Fetch a sample file path from DASCore
file_path = fetch('terra15_das_1_trimmed.hdf5')
folder_path = file_path.parent

spool = dc.spool(file_path)
# Print the spool object to see its contents
print(spool)
spool.get_contents()
````

````python
# for a folder
folder_path = file_path.parent
spool = dc.spool(folder_path)

# Print the spool object to see its contents
print(spool)
spool.get_contents()
````

## Writing DAS data files

Write to a file using the `patch.io.write()` method.

```python
import dascore as dc

patch = dc.get_example_patch()
patch.io.write("path/to/output/file", output_format)
```

- Example:

```python
patch = dc.get_example_patch()
patch.io.write("output_file.h5", 'dasdae')
```

````python
import dascore as dc

patch = dc.get_example_patch()
patch.io.write("output_file.h5", 'dasdae')
````

## Simple processing of DAS data

```python
import dascore as dc

patch = dc.get_example_patch('random_das')
# one processing step at a time
out = patch.pass_filter(time=(..., 10))
# Or chain processing steps together
out = (
    # Decimate along time axis (keep every 8th sample)
    patch.decimate(time=8)
    # Detrend along the distance dimension
    .detrend(dim='distance') 
    # Apply 10Hz low-pass filter along time dimension
    .pass_filter(time=(..., 10))
)
```

````python
import dascore as dc

patch = dc.get_example_patch('random_das')
# one processing step at a time
out = patch.pass_filter(time=(..., 10))
# Or chain processing steps together
out = (
    # Decimate along time axis (keep every 8th sample)
    patch.decimate(time=8)
    # Detrend along the distance dimension
    .detrend(dim='distance') 
    # Apply 10Hz low-pass filter along time dimension
    .pass_filter(time=(..., 10))
)
````

````python
patch.viz.waterfall()
out.viz.waterfall()
````
